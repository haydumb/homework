```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\17135\Downloads\archive\diamonds.csv', index_col=0)
# Print the top 7 records in the dataframe
print(df.head(7))
# Print the bottom 7 records in the dataframe
print(df.tail(7))
# Selecting the columns you want to describe
columns_to_describe = ["carat", "depth", "table", "price", "x", "y", "z"]
# Using the describe method on those specific columns
description = df[columns_to_describe].describe()
print(description)
# Using the Series method
series_to_describe = ["cut", "color", "clarity"]
for column in series_to_describe:
    description = df[column].describe()
    print(f"Descriptive statistics for {column}:")
    print(description)
    print()  # Adding a blank line for separation
# Get the unique values from the 'cut' column
unique_values = df['cut'].unique()
print("Unique category values in 'cut' column:", unique_values)
# Create the histogram
df.hist(figsize=(15, 10), bins=20)
plt.show()
```
